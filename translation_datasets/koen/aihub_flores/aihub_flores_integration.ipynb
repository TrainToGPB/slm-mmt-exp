```python
from comet import download_model, load_from_checkpoint
```

```python
model_path = download_model('Unbabel/XCOMET-XL')
model = load_from_checkpoint(model_path)
```

```python
data = [{"src": src_text, "mt": ref_text} for src_text, ref_text in zip(aihub_train['en'], aihub_train['ko'])]
data[:3]
```

```python
model_output = model.predict(data, batch_size=24, gpus=1)
print(model_output.scores)
```

```python
scores = [score * 100 for score in model_output.scores]
scores[:3]
```

```python
aihub_train['score'] = scores
aihub_train.head()
```

```python
aihub_train.columns = ['ko', 'en', 'source', 'xcomet']
aihub_train = aihub_train[['xcomet', 'ko', 'en', 'source']]
aihub_train.head()
```

```python
aihub_train.to_csv('aihub_train_with_score.csv', index=False)
```

```python
# xcomet 기준으로 aihub_train 정렬
aihub_train = aihub_train.sort_values(by='xcomet', ascending=True)
aihub_train.head(30)
```

```python
# 'source' 별 xcomet score 평균
aihub_train.groupby('source')['xcomet'].mean()
```

```python
# aihub_train에서 xcomet이 높은 소수의 데이터만 추출해 데이터 제작
# 'source'가 111, 125, 126, 71382인 경우: xcomet이 높은 순서대로 5000개씩 추출
# 'source'가 124, 563, 71265, 71266인 경우: xcomet이 높은 순서대로 2500개씩 추출
sources = [111, 125, 126, 71382, 124, 563, 71265, 71266]
aihub_train_selected = pd.DataFrame()
for source in sources:
    if source in [111, 125, 126, 71382]:
        aihub_train_selected = pd.concat([aihub_train_selected, aihub_train[aihub_train['source'] == source].tail(5000)], ignore_index=True)
    else:
        aihub_train_selected = pd.concat([aihub_train_selected, aihub_train[aihub_train['source'] == source].tail(2500)], ignore_index=True)
aihub_train_selected
```

```python
aihub_train_selected.columns = ['ko_ref_xcomet', 'ko_ref', 'en_ref', 'source']
aihub_train_selected
```

```python
aihub_train_selected.to_csv('sparta-30k-train.csv', index=False)
```

```python
flores_split_train.columns = ['en_ref', 'ko_ref', 'source']
flores_split_train
```

```python
data = [{"src": src_text, "mt": ref_text} for src_text, ref_text in zip(flores_split_train['en_ref'], flores_split_train['ko_ref'])]
data[:3]
```

```python
model_output = model.predict(data, batch_size=24, gpus=1)
flores_scores = [score * 100 for score in model_output.scores]
flores_scores[:3]
```

```python
flores_split_train['ko_ref_xcomet'] = flores_scores
flores_split_train = flores_split_train[['ko_ref_xcomet', 'ko_ref', 'en_ref', 'source']]
flores_split_train
```

```python
mixed_train_selected = pd.concat([aihub_train_selected, flores_split_train], ignore_index=True)
mixed_train_selected
```

```python
mixed_train_selected = mixed_train_selected.sample(frac=1).reset_index(drop=True)
mixed_train_selected
```

```python
# 'source'가 int인 경우 str로 변환하고 앞에 'aihub-'를 prefix로 붙이기
# 'source'가 str인 경우 앞에 'flores-'를 prefix로 붙이기
mixed_train_selected['source'] = mixed_train_selected['source'].apply(lambda x: f'aihub-{x}' if type(x) == int else f'flores-{x}')
mixed_train_selected.head(20)
```

```python
mixed_train_selected.to_csv('sparta-30k-mixed-train.csv', index=False)
```
