```python
from pprint import pprint

import pandas as pd
from datasets import load_dataset
```

```python
# cs-en, de-en, is-en, ru-en, zh-en
dataset = load_dataset("haoranxu/ALMA-R-Preference", 'de-en')
display(dataset)
```

```python
sources = ['cs-en', 'de-en', 'is-en', 'ru-en', 'zh-en']
for src in sources:
    dataset = load_dataset("haoranxu/ALMA-R-Preference", src)
    dataset = dataset['train']['translation']
    df = pd.DataFrame(dataset)
    df.to_csv(f'./{src}.csv', index=False)
```

```python
for data in dataset['train']['translation']:
    if data['Delta'] != 0:
        print(data['Delta'])
        required_direction = data['required_directions']
        src_lang, tgt_lang = required_direction.split('-')
        print(f"[{required_direction}] ALMA: {data[f'alma_{tgt_lang}_kiwi_xcomet']:.4f} / GPT-4: {data[f'gpt4_{tgt_lang}_kiwi_xcomet']:.4f}")
```
